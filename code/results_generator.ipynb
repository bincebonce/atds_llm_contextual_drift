```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
# qwen-2.5-1.5B results
qwen = pd.read_csv('../results/Qwen-2.5-1.5B Temporal Results/summary.csv')

plt.plot(["Baseline", "Round 1", "Round 2", "Round 3", "Round 4", "Round 5"], qwen['hallucination'])
plt.ylabel("Hallucination Rate")
plt.xlabel("Semantic Drift Round")
plt.grid()
plt.title("Hallucination Rate versus Semantic Drift")
plt.xticks(rotation=45)
plt.show()
```
